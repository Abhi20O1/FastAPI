```python
df_feat = df.copy()
```

```python
# Feature 1: BMI
df_feat['bmi'] = df_feat['weight']/(df_feat['height']**2)
```

```python
def age_group(age):
    if age < 25:
        return 'young'
    elif age < 45:
        return 'adult'
    elif age < 60:
        return 'middle_aged'
    return 'senior'
```

```python
df_feat["age_group"] = df_feat['age'].apply(age_group)
```

```python
# Feature 3: Lifestyle Risk
def lifestyle_risk(row):
    if row["smoker"] and row["bmi"] > 30:
        return "high"
    elif row["smoker"] or row["bmi"] > 27:
        return "medium"
    else:
        return "low"
```
